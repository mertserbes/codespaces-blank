````python
from datasets import Dataset, DatasetDict
import pandas as pd

# 1. JSON'dan oku
df = pd.read_json('f16_full_dataset.json')

# 2. Veri Formatını Ayarla (Hard Negative ve Easy Negative'i tek listede birleştirmek de mümkün,
# ama burada MultipleNegativesRankingLoss için ayrı sütunlarda tutacağız)
# Trainer için sütun isimleri 'anchor', 'positive', 'negative' (hard), 'negative_2' (easy) olacak.
# Pandas ile sütun adını 'hard_negative' -> 'negative', 'easy_negative' -> 'negative_2' yapalım.
df = df.rename(columns={"hard_negative": "negative", "easy_negative": "negative_2"})

# 3. Dataset Objesine Çevir
full_dataset = Dataset.from_pandas(df)

# 4. Train / Test / Validation Split Yap
# Önce Train (%80) ve Geriye Kalan (%20) olarak ayır
train_testvalid = full_dataset.train_test_split(test_size=0.2, seed=42)

# Sonra Geriye Kalanı (%20) -> Validation (%10) ve Test (%10) olarak ikiye böl
test_valid = train_testvalid['test'].train_test_split(test_size=0.5, seed=42)

# Hepsini tek bir DatasetDict'te topla
dataset_splits = DatasetDict({
    'train': train_testvalid['train'],
    'validation': test_valid['train'], # Buradaki 'train' ismi split fonksiyonundan gelir, ama bizim için validation'dır.
    'test': test_valid['test']
})

print("Veri Seti Bölündü:")
print(dataset_splits)

````

````python
from sentence_transformers import (
    SentenceTransformer,
    SentenceTransformerTrainer,
    SentenceTransformerTrainingArguments,
    losses
)
from sentence_transformers.evaluation import InformationRetrievalEvaluator

# 1. Modeli Başlat
model_name = "nomic-ai/modernbert-embed-base"
model = SentenceTransformer(model_name)

# 2. Loss Fonksiyonu (Triplet + Easy Negative)
# MultipleNegativesRankingLoss; anchor, positive, negative ve negative_2'yi otomatik kullanır.
train_loss = losses.MultipleNegativesRankingLoss(model=model)

# 3. Evaluator (Validasyon için) Hazırla
# Evaluator, validasyon seti içindeki soruları (queries) ve cevapları (corpus) alıp test eder.
val_anchors = dataset_splits['validation']['anchor']
val_positives = dataset_splits['validation']['positive']

# Corpus: Tüm cevaplar (Pozitifler)
corpus = {str(i): p for i, p in enumerate(val_positives)}
# Queries: Sorular
queries = {str(i): q for i, q in enumerate(val_anchors)}
# Hangi soru hangi cevaba ait? (1'e 1 eşleşme)
relevant_docs = {str(i): {str(i)} for i in range(len(queries))}

evaluator = InformationRetrievalEvaluator(
    queries=queries,
    corpus=corpus,
    relevant_docs=relevant_docs,
    name="f16-val-evaluator",
    show_progress_bar=True
)

# 4. Eğitim Argümanları
args = SentenceTransformerTrainingArguments(
    output_dir="modernbert-f16-finetuned",
    num_train_epochs=5,             # Veri az olduğu için epoch sayısını artırabiliriz
    per_device_train_batch_size=8,  # Colab'da RAM şişmesin diye 8 veya 16
    learning_rate=2e-5,
    warmup_ratio=0.1,
    fp16=True,                      # GPU hızlandırma
    eval_strategy="epoch",   # Her epoch sonu validasyon yap
    save_strategy="epoch",
    logging_steps=10,
    load_best_model_at_end=True,    # En iyi validasyon skoruna sahip modeli yükle
    metric_for_best_model="eval_f16-val-evaluator_cosine_mrr@10" # Mean Reciprocal Rank (Sıralama başarısı)
)

# 5. Trainer Başlat
trainer = SentenceTransformerTrainer(
    model=model,
    args=args,
    train_dataset=dataset_splits['train'],
    eval_dataset=dataset_splits['validation'],
    loss=train_loss,
    evaluator=evaluator
)

# 6. Eğit
trainer.train()

# 7. Kaydet
model.save_pretrained("f16-final-model")
print("Model başarıyla eğitildi ve doğrulandı.")
````

### Data Prep
We now use the `Gemma-3` format for conversation style finetunes. We use [Maxime Labonne's FineTome-100k](https://huggingface.co/datasets/mlabonne/FineTome-100k) dataset in ShareGPT style. Gemma-3 renders multi turn conversations like below:

```
<bos><start_of_turn>user
Hello!<end_of_turn>
<start_of_turn>model
Hey there!<end_of_turn>
```

We use our `get_chat_template` function to get the correct chat template. We support `zephyr, chatml, mistral, llama, alpaca, vicuna, vicuna_old, phi3, llama3, phi4, qwen2.5, gemma3` and more.

We now have to apply the chat template for `Gemma-3` onto the conversations, and save it to `text`. We remove the `<bos>` token using removeprefix(`'<bos>'`) since we're finetuning. The Processor will add this token before training and the model expects only one.

````python
def formatting_prompts_func(examples):
   convos = examples["conversations"]
   texts = [tokenizer.apply_chat_template(convo, tokenize = False, add_generation_prompt = False).removeprefix('<bos>') for convo in convos]
   return { "text" : texts, }

dataset = dataset.map(formatting_prompts_func, batched = True)
````

### Train the model
Now let's train our model. We do 60 steps to speed things up, but you can set `num_train_epochs=1` for a full run, and turn off `max_steps=None`.

````python
from trl import SFTTrainer, SFTConfig
trainer = SFTTrainer(
    model = model,
    tokenizer = tokenizer,
    train_dataset = dataset,
    eval_dataset = None, # Can set up evaluation!
    args = SFTConfig(
        dataset_text_field = "text",
        per_device_train_batch_size = 1,
        gradient_accumulation_steps = 4, # Use GA to mimic batch size!
        warmup_steps = 5,
        # num_train_epochs = 1, # Set this for 1 full training run.
        max_steps = 60,
        learning_rate = 2e-4, # Reduce to 2e-5 for long training runs
        logging_steps = 1,
        optim = "adamw_8bit",
        weight_decay = 0.001,
        lr_scheduler_type = "linear",
        seed = 3407,
        report_to = "none", # Use TrackIO/WandB etc
    ),
)
````
